### Utility functions

```python
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap, BoundaryNorm

def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_labelled_scatter(X, y, class_labels):
    num_labels = len(class_labels)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    marker_array = ['o', '^', '*']
    color_array = ['#FFFF00', '#00AAFF', '#000000', '#FF00AA']
    cmap_bold = ListedColormap(color_array)
    bnorm = BoundaryNorm(np.arange(0, num_labels + 1, 1), ncolors=num_labels)
    plt.figure()

    plt.scatter(X[:, 0], X[:, 1], s=65, c=y, cmap=cmap_bold, norm = bnorm, alpha = 0.40, edgecolor='black', lw = 1)

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)

    h = []
    for c in range(0, num_labels):
        h.append(mpatches.Patch(color=color_array[c], label=class_labels[c]))
    plt.legend(handles=h)
    plt.show()

    
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    plt.scatter(non_cores[:, 0], non_cores[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
```

## K-means model

The K-Means algorithm is one of the fastest clustering algorithms, but also one of the simplest:
* First initialize $k$ centroids randomly: $k$ distinct instances are chosen randomly from the dataset and the centroids are placed at their locations.
* Repeat until convergence (i.e., until the centroids stop moving):
    * Assign each instance to the closest centroid.
    * Update the centroids to be the mean of the instances that are assigned to them.
    
    
The `KMeans` class applies an optimized algorithm by default. To get the original K-Means algorithm (for educational purposes only), you must set `init="random"`, `n_init=1`and `algorithm="full"`. 

See the details of these hyperparameters: 

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

### 1. A synthetic dataset

This example creates a synthetic dataset with make_blobs, then applies k-means to find 3 clusters, and plots the points in each cluster identified by a corresponding color.

#### Generate a dataset

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state = 10)

plot_data(X)


```

#### K-means model 

```python
from sklearn.cluster import KMeans

# create a K-means model
kmeans = KMeans(n_clusters=3, init='random', n_init=1, max_iter=10, algorithm='full')
```

```python
# fit the model
kmeans.fit(X)
```

```python
# plot the resulting clusters

plot_labelled_scatter(X, kmeans.labels_, ['C', 'C2', 'C3'])
```

## k-means vs. DBSCAN clustering

### 1. A simple dataset

#### Generate a dataset

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state = 9, n_samples = 25)

plot_data(X)
```

#### K-means model

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

plot_labelled_scatter(X, kmeans.labels_, ['C1', 'C2', 'C3'])
```

#### DBSCAN model

Read the description of the model parameters

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html

```python
from sklearn.cluster import DBSCAN


dbscan = DBSCAN(eps=2, min_samples=2)
dbscan.fit(X)

plot_labelled_scatter(X, dbscan.labels_+1, ['noise', 'C1', 'C2', 'C3'])
```

```python
final_results=pd.DataFrame(columns =["Method","Accuracy","Silhouette Score"])
```

Try Gaussian Mixture model

```python
from sklearn.mixture import GaussianMixture
model = GaussianMixture(n_components=2, random_state=42).fit(X_normalized)

y_pred = model.predict(X_normalized)
```

```python
acc=max(np.sum(y_pred!=y)/len(y),np.sum(y_pred==y)/len(y))
print('Accuracy= %f'%acc)
```

```python
score = metrics.silhouette_score(X_normalized, y_pred, 
                metric='euclidean', sample_size=len(X_normalized))
print('Silhouette scores= %f'%score)
```

```python
final_results.loc[len(final_results.index)] = ['Gaussian Mixture',acc,score]
```

KMeans

```python
num_clusters=range(2,20)
scores = np.zeros(len(num_clusters))
accuracies=np.zeros(len(num_clusters))

for ii,num_cluster in enumerate(num_clusters):
    kmeans = KMeans(n_clusters=num_cluster)
    kmeans.fit(X_normalized)
    
    # extract performance metrics
    score = metrics.silhouette_score(X_normalized, kmeans.labels_, 
                metric='euclidean', sample_size=len(X_normalized))
    acc=max(np.sum(kmeans.labels_==y)/len(y),np.sum(kmeans.labels_!=y)/len(y))
    
    #print('\n k = {:} --> Silhouette score = {:.4f}'.format(epsillon, score))
                    
    scores[ii]=score
    accuracies[ii]=acc
```

```python
plt.figure()
plt.bar(num_clusters, scores, width=0.7, color='k', align='center')
plt.xlabel('num_clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette score vs num_clusters')
```

```python
plt.figure()
plt.bar(num_clusters, accuracies, width=0.7, color='k', align='center')
plt.xlabel('num_clusters')
plt.ylabel('Accuracies Score')
plt.title('Accuracies vs num_clusters')
```

```python
final_results.loc[len(final_results.index)] =["KMeans",accuracies[0],scores[0]]
```

DB Scan

```python
eps_vals=np.linspace(0.2, 1.2, num=15)
scores = np.zeros(len(eps_vals))
accuracies=np.zeros(len(eps_vals))

for ii,epsillon in enumerate(eps_vals):
    # Train the model
    dbscan = DBSCAN(eps=epsillon, min_samples=5)
    dbscan.fit(X_normalized)
    
    # extract performance metrics
    score = metrics.silhouette_score(X_normalized, dbscan.labels_+1, 
                metric='euclidean', sample_size=len(X_normalized))
    acc=max(np.sum(dbscan.labels_+1==y)/len(y),np.sum(dbscan.labels_+1!=y)/len(y))
    
    #print('\n k = {:} --> Silhouette score = {:.4f}'.format(epsillon, score))
                    
    scores[ii]=score
    accuracies[ii]=acc
```

```python
plt.figure()
plt.bar(eps_vals, scores, width=0.05, color='k', align='center')
plt.xlabel('Epsillon Value')
plt.ylabel('Silhouette Score')
plt.title('Silhouette score vs epsilon')
```

```python
plt.figure()
plt.bar(eps_vals, accuracies, width=0.05, color='k', align='center')
plt.xlabel('Epsillon Value')
plt.ylabel('Accuracies Score')
plt.title('Accuracies vs epsilon')
```

```python
final_results.loc[len(final_results.index)] = ['DB Scan',max(accuracies),scores[accuracies==max(accuracies)][0]]
```
