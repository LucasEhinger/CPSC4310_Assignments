### Set up 

```python
# Python ≥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import pandas as pd
import os

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# to make this notebook's output stable across runs
np.random.seed(42)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "polynomialRegression"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)

os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

### Synthetic dataset for regression

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_friedman1.html

```python
# synthetic dataset for more complex regression
from sklearn.datasets import make_friedman1

X, y = make_friedman1(n_samples = 100, n_features = 6, random_state=0)

print(type(X), X.shape)
print(type(y), y.shape)
print(y)
print(X)
```

```python
# Complex regression problem with one input variable

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=[8,8])

ax1.scatter(X[:, 0], y, marker= 'x', s=10)
ax2.scatter(X[:, 1], y, marker= 'x', s=10)
ax3.scatter(X[:, 2], y, marker= 'x', s=10)
ax4.scatter(X[:, 3], y, marker= 'x', s=10)
ax5.scatter(X[:, 4], y, marker= 'x', s=10)
ax6.scatter(X[:, 5], y, marker= 'x', s=10)



plt.show()
```

### Linear regression with multivariate

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

linreg = LinearRegression().fit(X_train, y_train)

print('linear model coeff (w): {}'.format(linreg.coef_))
print('linear model intercept (b): {:.3f}'.format(linreg.intercept_))

print('R-squared score (training): {:.3f}'.format(linreg.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'.format(linreg.score(X_test, y_test)))
```

### Polynomial with Linear regression

Now we transform the original input data to add polynomial features up to degree 2 (quadratic)

For example, if an input sample is two dimensional and of the form [a, b], the degree-2 polynomial features are [1, a, b, a^2, ab, b^2].


https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html




```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state=0)

linreg_poly = LinearRegression().fit(X_train, y_train)

print('(poly deg 2) linear model coeff (w):\n{}'
     .format(linreg_poly.coef_))
print('(poly deg 2) linear model intercept (b): {:.3f}'
     .format(linreg_poly.intercept_))
print('(poly deg 2) R-squared score (training): {:.3f}'
     .format(linreg_poly.score(X_train, y_train)))
print('(poly deg 2) R-squared score (test): {:.3f}\n'
     .format(linreg_poly.score(X_test, y_test)))
```

### Ridge regression (regularized linear regression with L2 penalty)

- Addition of many polynomial features often leads to overfitting,
so we often use polynomial features in combination with regression that has a regularization penalty, like ridge regression.

```python
from sklearn.linear_model import Ridge

X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 0)

linreg_ridge = Ridge(alpha=1).fit(X_train, y_train)

print('(poly deg 2 + ridge) linear model coeff (w):\n{}'
     .format(linreg_ridge.coef_))
print('(poly deg 2 + ridge) linear model intercept (b): {:.3f}'
     .format(linreg_ridge.intercept_))
print('(poly deg 2 + ridge) R-squared score (training): {:.3f}'
     .format(linreg_ridge.score(X_train, y_train)))
print('(poly deg 2 + ridge) R-squared score (test): {:.3f}'
     .format(linreg_ridge.score(X_test, y_test)))
```

### Lasso regression (regularized linear regression with L1 penalty)


https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html

```python
from sklearn.linear_model import Lasso

X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 0)

linreg_lasso = Lasso(alpha=0.01).fit(X_train, y_train)

print('(poly deg 2 + lasso) linear model coeff (w):\n{}'
     .format(linreg_lasso.coef_))
print('(poly deg 2 + lasso) linear model intercept (b): {:.3f}'
     .format(linreg_lasso.intercept_))
print('(poly deg 2 + lasso) R-squared score (training): {:.3f}'
     .format(linreg_lasso.score(X_train, y_train)))
print('(poly deg 2 + lasso) R-squared score (test): {:.3f}'
     .format(linreg_lasso.score(X_test, y_test)))
```

### Polynomial regression exercises

### 1. create a set of polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=10, include_bias=False)

X_poly = poly_features.fit_transform(X)
```

### 2. check values of X[0] and X_poly[0]

```python
print(X[0])

print(X_poly[0])
```

### 3. Linear regression modeling with data polynomial attributes

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 0)

linreg_poly = LinearRegression().fit(X_train, y_train)

print(linreg_poly.intercept_, linreg_poly.coef_)

print('R-squared score (training): {:.3f}'
     .format(linreg_poly.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linreg_poly.score(X_test, y_test)))
```

#### Normalize the data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

linreg_poly_norm = LinearRegression().fit(X_train_scaled, y_train)

print(linreg_poly_norm.intercept_, linreg_poly_norm.coef_)

print('(poly deg 2 Normalized) R-squared score (training): {:.3f}'
     .format(linreg_poly_norm.score(X_train_scaled, y_train)))
print('(poly deg 2 Normalized) R-squared score (test): {:.3f}\n'
     .format(linreg_poly_norm.score(X_test_scaled, y_test)))
```

### 4. Draw regression line

```python
# new data
X_new = np.linspace(-3, 3, 100).reshape(100, 1)

# polynomial attributes from testing data
X_test_poly = poly_features.fit_transform(X_new)

# predictve values
y_predict = linreg_poly.predict(X_test_poly)

# plot training data
plt.plot(X, y, "b.")

# plot regression line
plt.plot(X_new, y_predict, "r-", linewidth=2, label="degree2")

plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-3, 3, 0, 10])
save_fig("quadratic_predictions_plot")
plt.show()
```

### 5. Learning curves

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    
    train_errors, val_errors = [], []
    
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   # not shown in the book
    plt.xlabel("Training set size", fontsize=14) # not shown
    plt.ylabel("RMSE", fontsize=14)              # not shown
```

### Linear regression learning curve

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)

plt.axis([0, 80, 0, 3])                         # not shown in the book
save_fig("underfitting_learning_curves_plot")   # not shown
plt.show()                                      # not shown
```

### Polynomial regression learning curve

```python
poly = PolynomialFeatures(degree=5, include_bias=False)
X_poly = poly.fit_transform(X)

linreg_poly = LinearRegression()

plot_learning_curves(linreg_poly, X_poly, y)


plt.axis([0, 80, 0, 3])           # not shown
save_fig("learning_curves_plot")  # not shown
plt.show()                        # not shown
```

```python
linreg_ridge = Ridge(alpha=1)

plot_learning_curves(linreg_ridge, X_poly, y)


plt.axis([0, 80, 0, 3])                         # not shown in the book
save_fig("underfitting_learning_curves_plot")   # not shown
plt.show()                                      # not shown
```

```python
linreg_lasso = Lasso(alpha=0.1)

plot_learning_curves(linreg_lasso, X_poly, y)


plt.axis([0, 80, 0, 3])                         # not shown in the book
save_fig("underfitting_learning_curves_plot")   # not shown
plt.show()                                      # not shown
```

### High degree polynomials

- degree=1 is just linear regression

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

for style, width, degree in (("g-", 1, 100), ("b--", 2, 2), ("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
            ("poly_features", polybig_features),
            ("std_scaler", std_scaler),
            ("lin_reg", lin_reg),
        ])
    
    polynomial_regression.fit(X, y)
    y_predict = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_predict, style, label=str(degree), linewidth=width)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
save_fig("high_degree_polynomials_plot")
plt.show()
```
